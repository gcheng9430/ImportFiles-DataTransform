```python
# activity_df = pd.read_excel(user_input)
# mapping_df = pd.read_excel('Mapping/content/Cash_Rec_Mapping.xlsx')

activity_df = pd.read_excel('/content/Settled Cash Activity Statement - Vidrio_22 Jul 2022.xls') 
mapping_df = pd.read_excel('/content/Cash_Rec_Mapping.xlsx')
```

```python
activity_df = activity_df.fillna('')
```

```python
exception_df = pd.DataFrame()
```

```python
hasException = False
```

```python
activity_df['Bank Reference ID'] = activity_df['Reference Number']
```

```python
activity_df['Post Date'] = pd.to_datetime(activity_df['Cash Post Date'])
```

```python
activity_df['Value Date'] = pd.to_datetime(activity_df['Cash Value Date'])
```

```python
activity_df['Amount'] = activity_df['Transaction Amount Local']
```

```python
activity_df['Description'] = activity_df[["Transaction Description 1","Transaction Description 1","Transaction Description 1","Transaction Description 1","Transaction Description 1","Transaction Description 1","Detailed Transaction Type Name","Transaction Type"]].apply(
    lambda x: ','.join(x.dropna().astype(str)),
    axis=1
)
```

```python
activity_df['Bank Account'] = activity_df['Cash Account Number']
```

```python
activity_df['Closing_Balance'] = activity_df['Closing Balance Local']
```

```python
activity_df['Filename'] = activity_df['Cash Account Number'].apply(lambda x: str(x)+datetime.now().strftime(" %Y-%m-%d-%H-%M-%S")+'.csv')
```

```python
bank_id_df = pd.DataFrame(mapping_df['Bank Ref ID'])
```

```python
bank_id_balance_df = pd.DataFrame(mapping_df[['Bank Ref ID','Starting_Balance']])
```

```python
for bank_ref_id in bank_id_df['Bank Ref ID']:
  startinig_balance = bank_id_balance_df.loc[bank_id_balance_df['Bank Ref ID'] == bank_ref_id,'Starting_Balance']
  output_df = activity_df.loc[activity_df['Cash Account Number']== bank_ref_id]
  mm_df = output_df[output_df['Description'].str.contains('STIF')]
  output_df = output_df[~output_df['Description'].str.contains('STIF')]
  write_file_df = output_df[['Bank Reference ID', 'Post Date','Value Date','Amount','Description','Bank Account','Closing_Balance']]
  write_file_df = write_file_df.dropna(subset = ['Bank Reference ID'])
  if write_file_df.shape[0] == 0:
    print( bank_ref_id,'has no activity')
  else:
    bank_closing_balance = sum(activity_df.loc[activity_df['Bank Account']==bank_ref_id,'Closing_Balance'])
    mm_investment = sum(mm_df['Amount'])
    new_row = {'Bank Reference ID': "Starting Balance", 'Post Date':'2020-01-01','Value Date':'2020-01-01','Amount':startinig_balance,'Description':"Starting Balance",'Bank Account':bank_ref_id,'Closing_Balance':0}
    write_file_df = write_file_df.append(new_row, ignore_index = True)
    calc_closing_balance = sum(write_file_df['Amount'].apply(lambda x: x if isinstance(x, float) else 0))


    if mm_investment+bank_closing_balance != calc_closing_balance:
      exception_row = {'Bank ReferenceId':bank_ref_id,'Bank Closing Balance': bank_closing_balance,'MM Value':mm_investment,'Calculated Closing Balance':calc_closing_balance}
      exception_df.append(exception_row,ignore_index = True)
      hasException = True
    

    now = str(bank_ref_id)+ datetime.now().strftime(" %Y-%m-%d-%H-%M-%S") 
    address_file = sys.path[0] 
    address_file = f'{address_file}/Output/'
    # writer = pd.ExcelWriter(address_file, engine='openpyxl')
    write_file_df.to_excel((f'{address_file}{now}.xlsx'),'Bank Transactions', index=False) #Saving the file inside the subfolder via the stored path

    #new closing balance
    mapping_df[mapping_df['Bank Ref ID'] ==bank_ref_id]['Closing_Balance'] = calc_closing_balance


address_file = sys.path[0] 
address_file = f'{address_file}/Mapping/'
# writer = pd.ExcelWriter(address_file, engine='openpyxl')
write_file_df.to_excel((f'{address_file}Cash_Rec_Mapping.xlsx'), index=False) #Saving the file inside the subfolder via the stored path
if hasException:
  address_file = sys.path[0] 
  output_path = f'{address_file}/Output/'
  # writer = pd.ExcelWriter(address_file, engine='openpyxl')
  exception_df.to_excel((f'{output_path}Exceptions.xlsx'), index=False)






```
